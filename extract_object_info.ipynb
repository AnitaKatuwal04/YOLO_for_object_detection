```python
import os 
from glob import glob 
import pandas as pd
from xml.etree import ElementTree as Et
from functools import reduce
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
#step1: Get path of each xml file
xmlfiles=glob('./images1/*.xml')
```

```python
xmlfiles
```

```python
#step2-read xml files
def extract_text(filename):
    tree=Et.parse(filename)
    root=tree.getroot()
    
    #extract filename
    image_name=root.find('filename').text
    #width and height of the image
    width=root.find('size').find('width').text
    height=root.find('size').find('height').text
    objs=root.findall('object')
    parser=[]
    for obj in objs:
        name=obj.find('name').text
        bndbox=obj.find('bndbox')
        xmin=bndbox.find('xmin').text
        xmax=bndbox.find('xmax').text
        ymin=bndbox.find('ymin').text
        ymax=bndbox.find('ymax').text
        parser.append([image_name,width,height,name,xmax,xmin,ymax,ymin])
        
    return parser
```

```python
parser_all=list(map(extract_text,xmlfiles))
parser_all
```

```python
data=reduce(lambda x,y:x+y,parser_all)
data
```

```python
df=pd.DataFrame(data,columns=['image_name','width','height','name','xmax','xmin','ymax','ymin'])
df.head()
```

```python
df['name'].value_counts()
```

```python
#typeconversion
cols=['width','height','xmax','xmin','ymax','ymin']
df[cols]=df[cols].astype(int)
df.info()
```

```python
#center x, center y
df['center_x']=((df['xmin']+df['xmax']/2)/df['width'])
df['center_y']=((df['ymin']+df['ymax']/2)/df['height'])
#W
df['w']=((df['xmax']-df['xmin'])/df['width'])
df['h']=((df['ymax']-df['ymin'])/df['height'])
```

```python
df.head()
```

```python
#split into train and test
images=df['image_name'].unique()
```

```python
len(images)
```

```python
#80% train and 20% test
img_df=pd.DataFrame(images,columns=['image_name'])
img_train=tuple(img_df.sample(frac=0.8)['image_name'])
```

```python
img_test=tuple(img_df.query(f'image_name not in {img_train}')['image_name'])
```

```python
train_df=df.query(f'image_name in {img_train}')
test_df=df.query(f'image_name in {img_test}')
```

```python
#Do encoding
def label_encoding(x):
    labels = {'toy':0, 'butterfly':1, 'cup':2, 'flower':3, 'helicopter':4,
              'shoe':5, 'dog':6, 'cat':7, 'car':8,'chair':9, 'bird':10}
    try:
        return labels[x]
    except KeyError:
        # Handle cases where 'x' is not found in the labels dictionary
        return -1  # Or any other appropriate value or action
```

```python

df['labels']=df['name'].apply(label_encoding)
```

```python
df.head()
```

```python
train_df['id']=train_df['name'].apply(label_encoding)
test_df['id']=train_df['name'].apply(label_encoding)
```

```python
cols=['image_name','id','center_x','center_y','w','h']
groupby_obj_train = train_df[cols].groupby('image_name')
groupby_obj_test = test_df[cols].groupby('image_name')
```

```python
import shutil
def save_data(image_name, folder_path, group_obj):
    src_path=os.path.join('images', image_name)
    dst_path=os.path.join(folder_path, image_name)
    
    #Move the image file
    if os.path.exists(src_path):
        shutil.move(src_path, dst_path)
    else:
        print("file not found")
    
    #save the labels
    text_filename=os.path.join(folder_path, os.path.splitext(image_name)[0]+'.txt')
    group_data=group_obj.get_group(image_name).set_index('image_name')
    group_data.to_csv(text_filename, sep=' ', index=False, header=False)
    
#ensure the directories exists
train_folder='images/train'
test_folder='images/test'
os.makedirs(train_folder, exist_ok=True)
os.makedirs(test_folder, exist_ok=True)

#Apply the functions to training filenames
filename_series_train=pd.Series(groupby_obj_train.groups.keys())
filename_series_train.apply(lambda image_name: save_data(image_name, train_folder, groupby_obj_train))

filename_series_test=pd.Series(groupby_obj_test.groups.keys())
filename_series_test.apply(lambda image_name: save_data(image_name, test_folder, groupby_obj_test))

    
```
